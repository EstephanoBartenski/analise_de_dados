```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips['tip'].plot(kind='hist')
```

```python
cts = tips['smoker'].value_counts()
cts
```

```python
cts.plot(kind='bar')
```

```python
sns.countplot(x='smoker', data=tips)
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False, col='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False, col='smoker', row='time')
```

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex', palette=['coral', 'indianred'])
facet.map(plt.scatter, 'total_bill', 'tip')
```

```python
fig, (ax) = plt.subplots(nrows=1, ncols=1)
```

```python
fig, (ax) = plt.subplots(nrows=1, ncols=1)
ax.scatter(x=tips['tip'], y=tips['total_bill'])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(tips['tip'], tips['total_bill'])
ax2.hist(tips['total_bill'])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
sns.distplot(tips['tip'], color='tan', ax=ax1)
sns.regplot(x='total_bill', y='tip', data=tips, color='coral', ax=ax2)
```
